```python
import matplotlib.pyplot as plt

num_threads = []
speedup_v1 = []
speedup_v2 = []
speedup_v3 = []

with open('log_bigram.txt', 'r') as file: # or use the log_trigram.txt file
    for line in file:
        if line.startswith('NUM THREADS'):
            num_threads.append(int(line.split()[-1]))
        elif line.startswith('PARALLEL v1'):
            speedup_v1.append(float(line.split(':')[-1].strip()))
        elif line.startswith('PARALLEL v2'):
            speedup_v2.append(float(line.split(':')[-1].strip()))
        elif line.startswith('PARALLEL v3'):
            speedup_v3.append(float(line.split(':')[-1].strip()))

plt.plot(num_threads, speedup_v1, label='Parallel v1', color='red')
plt.plot(num_threads, speedup_v2, label='Parallel v2', color='blue')
plt.plot(num_threads, speedup_v3, label='Parallel v3', color='green')

plt.scatter(num_threads, speedup_v1, color='red')
plt.scatter(num_threads, speedup_v2, color='blue')
plt.scatter(num_threads, speedup_v3, color='green')

plt.legend()
plt.xlabel('Number of threads')
plt.ylabel('Speedup')
plt.title('Speedup trend varying the number of threads')

plt.grid(True)
plt.show()
```

```python
num_texts = []
speedup_v1 = []
speedup_v2 = []
speedup_v3 = []

with open('log_bigram_size.txt', 'r') as file: # or use the log_trigram_size.txt file
    for line in file:
        if line.startswith('Texts size'):
            # Rimuovi "MB" dal valore e converti in intero
            num_texts.append(int(line.split()[-2]))
        elif line.startswith('PARALLEL v1'):
            speedup_v1.append(float(line.split(':')[-1].strip()))
        elif line.startswith('PARALLEL v2'):
            speedup_v2.append(float(line.split(':')[-1].strip()))
        elif line.startswith('PARALLEL v3'):
            speedup_v3.append(float(line.split(':')[-1].strip()))

plt.plot(num_texts, speedup_v1, label='Parallel v1', color='red')
plt.plot(num_texts, speedup_v2, label='Parallel v2', color='blue')
plt.plot(num_texts, speedup_v3, label='Parallel v3', color='green')

plt.scatter(num_texts, speedup_v1, color='red')
plt.scatter(num_texts, speedup_v2, color='blue')
plt.scatter(num_texts, speedup_v3, color='green')

plt.legend()
plt.xlabel('Total size of texts [MB]')
plt.ylabel('Speedup')
plt.title('Speedup trend varying the number of threads')

plt.grid(True)
plt.show()
```
